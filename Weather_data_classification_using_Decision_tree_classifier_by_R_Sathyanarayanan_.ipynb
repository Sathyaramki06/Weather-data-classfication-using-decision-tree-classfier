```python
weather= weather.dropna() #all the null values will be dropped off
```

```python
data_clean= weather.copy()
data_clean
```

```python
data_clean['high humidity']=(data_clean['relative_humidity_3pm'] >24.99)*1
data_clean['high humidity'].head()
```

```python
y=data_clean['high humidity'].copy()
y.head()
```

```python
data_clean.head()
```

```python
# creating a new data set for morning features

morning_features=['air_pressure_9am', 'air_temp_9am', 'avg_wind_direction_9am',
       'avg_wind_speed_9am', 'max_wind_direction_9am', 'max_wind_speed_9am',
       'rain_accumulation_9am', 'rain_duration_9am', 'relative_humidity_9am']
```

```python
x=data_clean[morning_features].copy()
x.columns
```

```python
y= y.to_frame() #converting the series to data frame
```

```python
type(y)
```

```python
y.columns
```

```python
# Now we gonna perform the train teat split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.33)
```

```python
# now we can check the shape of x train and y train
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
#now we gonna train the model
humidity_classifier=DecisionTreeClassifier(max_leaf_nodes=10,random_state=0)
humidity_classifier.fit(X_train,y_train)
```

```python
type(humidity_classifier)
```

```python
# Now lets predict on the test set
y_prediction=humidity_classifier.predict(X_test)
```

```python
y_prediction[:10]
```

```python
y_test[:10]
```

```python
accuracy_score(y_test,y_prediction)*100
```
